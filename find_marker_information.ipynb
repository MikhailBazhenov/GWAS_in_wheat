```python
import pandas as pd
```

```python
mta = pd.read_excel('Significant_MTAs.xlsx', sheet_name='Significant MTA')
markers = mta['Marker'].to_list()
markers
```

```python
snp_60k = pd.read_excel('60K_SNP.xlsx', sheet_name='Total_60K_SNP', index_col='Identifier')
snp_60k.head()
```

```python
genotypes = pd.read_excel('Genotypes_filtered_imputed.xlsx', index_col='rs#')
genotypes.head()
```

```python
output = pd.DataFrame(columns=['marker', 'ref', 'alt', 'description', 'type'])
```

```python
for i, marker in enumerate(markers):
    if marker == 'IAB60001':
        output.loc[i] = [marker, 'ref', 'alt', 'description', 'type']
        continue
    ref = snp_60k['Reference allele'][marker]
    alt_variants = genotypes['alleles'][marker].split('/')
    for variant in alt_variants:
        if variant != ref:
            alt = variant
    description = snp_60k['Description'][marker]
    type = snp_60k['Type'][marker]
    output.loc[i] = [marker, ref, alt, description, type]

output
```

```python
output.to_excel('found_marker_information.xlsx')
```
