```python
# Preparation of numerical genotypes table got from TASSEL for processing with SimpleM R script
# (for calculation the number of effective markers and avoiding high Bonferroni penalty)
# See: https://doi.org/10.1002/gepi.20430

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('numerical_genotypes.txt', sep='\t', decimal=',', skiprows=1)
df.set_index(df['<Marker>'], inplace=True)
df.drop(columns='<Marker>', inplace=True)
df.head()
```

```python
array = np.array(df, dtype=int)
array = array * 2
array = array.transpose()
```

```python
df_out = pd.DataFrame(array)
df_out
```

```python
df_out.to_csv('genotypes.txt', sep=' ', header=False, index=False)
```
