```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Genotypes_seqSNP+KASP.hmp.txt', sep='\t')

columns = df.columns.to_list()
cultivars = columns[11:]
df['missing'] = (df[cultivars] == 'N').sum(axis=1) / len(cultivars)
df_missing = df[[columns[2], columns[3], 'missing']]
df_missing.set_index('chrom', inplace=True)

for chrom in df_missing.index.unique():
    df_chrom = df_missing.loc[chrom]
    plt.figure(chrom, figsize=(4, 2)) 
    plt.scatter(df_chrom['pos'], df_chrom['missing'], alpha=0.2)
    plt.title('Chromosome ' + chrom)
    plt.ylabel('No-call SNP proportion')
    plt.xlabel('Position on chromosome, bp')
    plt.savefig(chrom + '_16.5K.png', bbox_inches='tight', dpi=400)
    plt.show() 
```
