```python
import pandas as pd
```

```python
snp_60k = pd.read_excel('60K_SNP.xlsx', sheet_name='Total_60K_SNP', index_col='Identifier')
snp_60k.head()
```

```python
df_out = pd.DataFrame()
```

```python
file = r'Genotypes_seqSNP+KASP_16495sites_90%call_hz10%max_maf5%min.hmp.txt'
df1 = pd.read_csv(file, sep='\t', index_col='rs#')
column = file.split('.')[0][22:]
df1['Type'] = snp_60k['Type']
df_out[column] = df1.groupby(['Type']).agg({'Type':'count'})
df_out.loc['sum'] = {column: df_out.loc[['A', 'B', 'C', 'D', 'G']][column].sum()}
df_out
```

```python
df_marker_by_chr = pd.DataFrame()
```

```python
df_marker_by_chr['60k'] = (
    snp_60k
    .groupby(['Reference'])
    .agg({'Reference': 'count'})
    )
df_marker_by_chr.head()
```

```python
df4['chrom_new'] = 'chr' + df4['chrom'].replace('UN', 'Un')
df_marker_by_chr['16k'] = (
    df4
    .groupby(['chrom_new'])
    .agg({'chrom_new': 'count'})
    )
df_marker_by_chr.tail()
```

```python
df3['chrom_new'] = 'chr' + df3['chrom'].replace('UN', 'Un')
df_marker_by_chr['9k'] = (
    df3
    .groupby(['chrom_new'])
    .agg({'chrom_new': 'count'})
    )
df_marker_by_chr.tail()
```

```python
df2['chrom_new'] = 'chr' + df2['chrom'].replace('UN', 'Un')
df_marker_by_chr['7k'] = (
    df2
    .groupby(['chrom_new'])
    .agg({'chrom_new': 'count'})
    )
df_marker_by_chr.tail()
```

```python
df1['chrom_new'] = 'chr' + df1['chrom'].replace('UN', 'Un')
df_marker_by_chr['3k'] = (
    df1
    .groupby(['chrom_new'])
    .agg({'chrom_new': 'count'})
    )
df_marker_by_chr.tail()
```

```python
df_marker_by_chr.to_excel('number_of_markers_on_chromosomes.xlsx')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_marker_by_chr.index.name = 'Chromosome'
```

```python
df_marker_by_chr['chr'] = df_marker_by_chr.index.str[3:]
```

```python
for column in df_marker_by_chr.columns:
    if column == 'chr':
        continue
    plt.figure(figsize=(7, 3), dpi=400)
    ax = sns.barplot(data=df_marker_by_chr, x='chr', y=column, palette='muted')
    ax.set_xlabel('Chromosome')
    ax.set_ylabel('Number of markers (' + column + '-set)')
    plt.savefig('Number of markers (' + column + '-set).png', bbox_inches='tight')
    plt.show()
```
